## Define the Following:
You can add image, diagrams, whatever you need to ensure that you understand the concepts below.

### Input Layer:
The inputs to the network. This is the layer that takes the features of your dataset into the model. The rest of the model's nodes can be thought of as a function that creates the output node values
### Hidden Layer:
All the layers inbetween the input and output layer. The structure of the hidden layers determines the "Effective function" that the neural network models
### Output Layer:
The Output of the network. The target. The value you guess based on input values
### Neuron:
A single node in a neural network. It has a function that defines how it processes data internally. For a hidden layer this is a combination of a weighted sum and a sigmoid function ( to make sure the influence of any one part of the node graph isn't higher than any other part of the node graph.
### Weight:
A modification of a value based on its importance in the network. Weights are initialized randomly, and then they are tuned in the training process.
### Activation Function:
A function that maps an input value ( often a weighted sum of the previous layer of the neural net ) to an output range between 0 and 1 
### Node Map:
A visual of the structure of the neural network
### Perceptron:
A simple neural network with just one input layer, and one output layer.

```python
##### Update this Class #####

class Perceptron(object):
    
    def __init__(self, niter = 10):
        self.niter = niter
    
    def sigmoid(self, x):
        return 1/(1+np.exp(-x))
    
    def sigmoid_derivative(self, x):
        dx = self.sigmoid(x)
        return dx*(1-dx)

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [#samples, #features]
        y : Target values, y.shape : [#samples]
        """
        # Randomly Initialize Weight
        # this is with the assumption that the perceptron
        # had two input neurons.
        
        # Initialize n variables reshaped to -1,1 (column vector) 
        num_weights = X.shape[1]
        weights = np.random.randn(num_weights).reshape(-1,1)
        print(weights)
        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X,weights)
            # Activate!
            activated_output = self.sigmoid(weighted_sum)
#             print('Activated output',activated_output)
            # Calc error
            error = y - activated_output
#             print("Error",error)
            adjustments = error * self.sigmoid_derivative(activated_output)
            # Update the Weights
            weights += np.dot(X.T,adjustments)
#             print("Weights",weights)
        self.weights = weights

    def predict(self, X):
        """Return class label after unit step"""
        print("Weights after training")
        crprint([self.weights.shape,self.weights])
        return self.sigmoid(np.dot(X,self.weights))
```
